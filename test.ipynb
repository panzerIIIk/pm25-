### 存SQLlite 

```python
import sqlite3
```

```python
# unique 插入資料唯一的約束
sqlstr='''
create table if not exists data(
id integer primary key autoincrement,
site text,
county text,
pm25 integer,	
datacreationdate text,	
itemunit text,
unique(site,datacreationdate)
)
'''
```

```python
conn =sqlite3.connect("pm25.db")
cusor=conn.cursor()

conn,cusor
```

```python
cusor.execute(sqlstr)
conn.commit()
```

### 插入資料
- or ignore 忽略

```python
sqlstr='insert or ignore into data (site,county,pm25,datacreationdate,itemunit)\
    values(?,?,?,?,?)'
    
```

```python
cusor.executemany(sqlstr,df1.values.tolist())
conn.commit()
```
